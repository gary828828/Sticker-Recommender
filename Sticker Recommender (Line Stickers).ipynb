```python
import requests
import bs4 as bs
import re
import shutil
import os
import time
```

```python
def download_stickers(img_urls):

    path = 'Line Stickers'
    for idx, img_url in enumerate(img_urls): 
        #enumarte to create a file integer name for every image
    
        # make a request to the image URL
        print("Downloading Sticker " + str(idx))
        img_source = requests.get(img_url, stream=True) 
        # we set stream = True to download/ 
        # stream the content of the data
        with open(path + os.sep + 'img'+str(idx)+'.jpg', 'wb') as file: 
            # open file connection, create file and write to it
            shutil.copyfileobj(img_source.raw, file) 
            # save the raw file object
    
        del img_source # to remove the file from memory
```

```python
#scrap stickers packages name of each page(url)
sticker_list = []

#insert page number(begin from 1 to 25)
for i in range(1,26):
    
    print('Downloading Stickers on page' + str(i))
    url = 'https://store.line.me/stickershop/showcase/top/en?page=' + str(i)
    
    raw = requests.get(url).content
    soup = bs.BeautifulSoup(raw,features='html.parser')
    
    #scrap stickers packages link of each page(url_0,url_list_0)
    url_list_0 = []
    for content in soup.find_all('img'): 
        url_0 = content.get('src') 
        if '.jpg' in url_0 or '.jpeg' in url_0 or '.png' in url_0:
            url_list_0.append(url_0)
    print('Number of sticker packages: ',len(url_list_0))
    
    #scrap stickers packages code of each page
    code = []
    for url_0 in url_list_0:
        result = re.search('https://stickershop.line-scdn.net/stickershop/v1/product/(.*)/LINEStorePC/thumbnail_shop.png;compress=true', url_0)
        code.append(result.group(1))
    
    #store stickers packages link of each page(url_1,url_list_1)
    url_list_1 = []     
    count = 1 
    for number in code:
        url_1 = 'https://store.line.me/stickershop/product/' + number +'/en'
        url_list_1.append(url_1)
        print('Loading Package ' + str(count))
            
        raw1 = requests.get(url_1).content
        soup1 = bs.BeautifulSoup(raw1,features='html.parser')
            
        #scrap stickers link of each sticker package(url_2,url_list_2)
        url_list_2 = []
        for content1 in soup1.find_all(attrs={"class": "mdCMN09Li FnStickerPreviewItem"}):
            url_2 = re.search(' \"staticUrl\" : \"(.*)\", \"fallbackStaticUrl\"', str(content1))
            sticker_list.append(url_2.group(1))
                
        print('Package ' + str(count) + ' Finish')
        count += 1
        
#            for url in url_list_2_raw:
#                url_2 = re.search(' \"staticUrl\" : \"(.*)\", \"fallbackStaticUrl\"', url)
#                sticker_list.append(url_2.group(1))
#            print('Package ' + str(count) + ' Finish')

            
```

```python
import pytesseract
from PIL import Image, ImageFile
import os
```

```python
#delete broken files(Usually the same size)
path="Line Stickers"
filelist = os.listdir(path)
print(len(filelist))
```

```python


print(len(filelist))
```

```python
filelist[0:10]
```

```python
image = Image.open(path + os.sep + 'img26849.jpg')
label = pytesseract.image_to_string(image, lang='eng')
```

```python
#character recognition
length = len(filelist)
n = 0
ImageFile.LOAD_TRUNCATED_IMAGES = True
for name in filelist:
    try:
        image = Image.open(path + os.sep + name)
        label = pytesseract.image_to_string(image, lang='eng')    
        length -= 1
        print(str(length) + " pics left")
        if label != '':
            n += 1
            print(str(n) + ' pics with label successfully detected')
            oldname= path + os.sep + name 
            newname= path + os.sep + label + '.jpg'
            os.rename(oldname,newname) 
            print(oldname,'======>',newname)
    except OSError:
        print('cannot open', name)
os.system('say "your program has finished"')
```
